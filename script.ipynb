```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
folder = '\data\Data - Survey Monkey Output.xlsx'
file = cwd + folder
```

```python
df = pd.read_excel(file, sheet_name='edited_table')
```

```python
df
```

```python
df.columns
```

```python
cols_to_drop = ['Start Date-', 'End Date-', 'Email Address-',
       'First Name-', 'Last Name-', 'Custom Data 1-',]
```

```python
df_input = df.copy()
df_input.drop(columns=cols_to_drop, inplace=True)
```

```python
df_input
```

```python
df_input.columns
id_vars = df_input.columns[:8]
```

```python
value_vars = df_input.columns[8:]
```

```python
df_melted = df_input.melt(id_vars=id_vars, value_vars=value_vars, value_name='Answers', var_name='Question-Subquestion')
df_melted
```

```python
df_melted['Question'] = df_melted['Question-Subquestion'].apply(lambda x: x.split('-')[0])
```

```python
df_melted
```

```python
df_melted.rename(columns=
    {
        'Respondent ID-': 'RespondentID'
    }, inplace=True
)
```

```python
filt = df_melted['Answers'].notnull()
responses = df_melted.loc[filt, :]
```

```python
questions_group = responses.groupby('Question')
```

```python
respondent_and_answers_count = questions_group['RespondentID'].agg(['nunique', 'count'])
```

```python
pd.merge(left= df_melted, right= respondent_and_answers_count, how= 'left', left_on= 'Question', right_on= 'Question')
```

```python
questions_answers = df_melted.copy()
questions_answers
```

```python
question_subquestion_group = questions_answers.groupby('Question-Subquestion')
```

```python
multi_index = question_subquestion_group['Answers'].value_counts(dropna=False)
multi_index
```

```python
multi_df = multi_index.unstack()
multi_df
```

```python
pd.merge(left= questions_answers, right= multi_index, how='left', left_on= ['Question-Subquestion', 'Answers'], right_index= True)
```
